#### 1. Encoding : converting binary to 0/1, Other categorical variable to dummy variable.

```python
# Dummy variable creation for month, weekday, weathersit, and season

month_df = pd.get_dummies(df.month, drop_first=True, dtype=int)
weekday_df = pd.get_dummies(df.weekday, drop_first=True, dtype=int)
weathersit_df = pd.get_dummies(df.weathersit, drop_first=True, dtype=int)
seasons_df = pd.get_dummies(df.season, drop_first=True, dtype=int)
```

```python
# Join the dataframe with the dummy variable dataset
df_new = pd.concat([df,month_df,weekday_df, weathersit_df,seasons_df], axis=1)
```

### Building the model with RFE

```python
# Building the liner model

y_train = df_train.pop('cnt')
X_train = df_train
```

```python
# Import RFE and LinearRegression 
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
lm_1 = LinearRegression()
lm_1.fit(X_train, y_train)

# Running RFE
rfe = RFE(lm_1, step=15)
rfe = rfe.fit(X_train, y_train)
```

```python
# Show the ranking of variable using RFE
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# Select the variable which RFE support True
col = X_train.columns[rfe.support_]
col
```

```python
# Show the columns which RFE did not slect
X_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

```python
# Create X_test dataframe with RFE selected variable
X_train_rfe = X_train[col]
```

```python
# import statsmodel
import statsmodels.api as sm

# Adding a constant variable
X_train_rfe = sm.add_constant(X_train_rfe)

# Running the linera model
lm_1 = sm.OLS(y_train, X_train_rfe).fit()



# show Summary
print(lm_1.summary())
```

#### Finding the VIF 

```python
# import VIF from statsmodel
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Function to calculate VIF of variables
def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
# Calculate VIF

calculateVIF(X_train_rfe)
```

#### Observations
- `atemp` has high p value. so it can be dropped
- `temp` , `atemp` have high VIF value

```python
# First Drop the 'atemp' column since it has high p value
X_train_new = X_train_rfe.drop(['atemp'], axis=1)
```

```python
# Adding constant variable
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train_new)

# Running the linear model
lm_2 = sm.OLS(y_train, X_train_lm).fit()

# show Summary
print(lm_2.summary())
```

```python
# Calculate VIF
calculateVIF(X_train_new)
```

```python
# Drop the 'spring' column since it has high VIF value
X_train_new = X_train_new.drop(['spring','dec'], axis=1)
```

```python
# Adding constant variable
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train_new)

# Running the linear model
lm_3 = sm.OLS(y_train, X_train_lm).fit()

# show Summary
print(lm_3.summary())
```

```python
# Calculate VIF
calculateVIF(X_train_new)
```

```python
X_train_new.columns
```

#### Observations 
- Now all variables have low p value(less than 0.05) and low VIF(less than 5), states that the overall model is significant

## Step 4 - Residual Analysis

#### R^2 value of test
